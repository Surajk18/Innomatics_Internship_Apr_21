Mean is an essential concept in mathematics and statistics. The mean is the average or the most common value in a collection of numbers. In statistics, it is a measure of central tendency of a probability distribution along median and mode. It is also referred to as an expected value.

Formula : $\frac{\Sigma x_i}{n}$ where $x_i$ is each value of dataset , n=total number of values in dataset

```python
col=data['Annual_HH_Income']
# mean without library
mean= sum(col)/len(col)
print(mean)
plt.bar(1,mean)
```

```python
# mean with library , used the statistics library

print(st.mean(col))
```

The median is the middle number in a sorted, ascending or descending, list of numbers and can be more descriptive of that data set than the average. ... If there is an odd amount of numbers, the median value is the number that is in the middle, with the same amount of numbers below and above.

n=number of values in dataset First sort the data in the increasing order to find the median value.

if n is odd ,then median = ($\frac{n}{2}$)th term

if n is even, then median = (($\frac{n}{2}$)th term +($\frac{n}{2}$+1)th term)/2

```python
# median without library

col=list(col)
col.sort()
n=len(col)
if n%2==0:
    median= (col[n//2]+col[(n//2)-1])/2
else:
    median=col[n//2]
    
print(median)

plt.bar(1,median)
```

```python
# median with library

print(st.median(col))
```

In statistics, the mode is the most commonly observed value in a set of data. For the normal distribution, the mode is also the same value as the mean and median. In many cases, the modal value will differ from the average value in the data.

If we have 2 or more numbers with the same frequency then all the numbers are said to be the modal values.



```python
# Mode without library

dic={}
for i in col:
    if i in dic:
        dic[i]+=1
    else:
        dic[i]=1
ma=0
for x in dic:
    if(ma<dic[x]):
        ma=dic[x]
        mode=x 
    
print("mode :",mode)
plt.bar(1,mode)
```

```python
# with library
print(st.mode(col))
```

variance is a measure of how spread out a data set is. It is calculated as the average squared deviation of each number from the mean of a data set.

There are two types of variance : i) Population variance :It is denoted by $\sigma^{2}$
ii) Sample variance : It is denoted by $S^{2}$
Formula for variance:

Population Variance($\sigma^{2}$) = $\frac{\Sigma_{i=1-n} (x_i-\mu)^{2}}{n}$
Sample variance( $S^{2}$)= $\frac{\Sigma_{i=1-n} (x_i-\bar{x})^{2}}{n-1}$



```python

#variance without library
su=0
for i in col:
    su+= (i-mean)**2
pop_var=su/len(col)
sam_var=su/(len(col)-1)
print("Poulation Variance :",pop_var)
print("Sample Variance :",sam_var)
plt.bar(1,pop_var)
```

```python

# Variance with library

print(st.pvariance(col))
print(st.variance(col))
```

A standard deviation is a statistic that measures the dispersion of a dataset relative to its mean. The standard deviation is calculated as the square root of variance by determining each data point's deviation relative to the mean.



```python
# Standard deviation without library
pop_std=pop_var**0.5
sam_std=sam_var**0.5

print("Population Standard Deviation :",pop_std)
print("Sample Standard Deviation :",sam_std)

plt.bar(1,pop_std)
```

```python
# standard Deviation using library

print("Population Standard Deviation :",st.pstdev(col))
print("Sample Standard Deviation :",st.stdev(col))
```

Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate). It's a common tool for describing simple relationships without making a statement about cause and effect.



```python
# Correlation without library
X=list(data['Mthly_HH_Expense'])
Y=col
mu_x=st.mean(X)
mu_y=st.mean(Y)
std_x=st.pstdev(X)
std_y=st.pstdev(Y)

su=0
for i in range(len(X)):
    su+= (X[i]-mu_x)*(Y[i]-mu_y)

su=su/(len(X)*std_x*std_y)
print("Correlation Coefficient :",su)
```

```python
# Correlation with library
print("Correlation Coefficient :",np.corrcoef(X,Y))
```

Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graph form, normal distribution will appear as a bell curve.



```python
normal=np.random.normal(Y)
sns.distplot(normal)
```

```python
normal_10000 = np.random.normal(loc = 20, scale = 5, size=15000)

sns.distplot(normal_10000)
```

```python
mean_normal= normal_10000.mean()

sigma = normal_10000.std()

one_std_right= mean_normal+sigma
one_std_left= mean_normal-sigma
two_std_right= mean_normal+(2*sigma)
two_std_left= mean_normal-(2*sigma)
three_std_right= mean_normal+(3*sigma)
three_std_left= mean_normal-(3*sigma)
```

```python
plt.figure(figsize=(20,10))
sns.set_style("darkgrid")
sns.distplot(normal_10000)

plt.axvline(mean_normal, color='coral', label='Mean')

plt.axvline(one_std_right, color='yellow', label='Mean + 1SD')
plt.axvline(one_std_left, color='yellow', label='Mean - 1SD')
plt.axvline(two_std_right, color='green', label='Mean + 2SD')
plt.axvline(two_std_left, color='green', label='Mean - 2SD')
plt.axvline(three_std_right, color='blue', label='Mean + 3SD')
plt.axvline(three_std_left, color='blue', label='Mean - 3SD')
plt.legend();
```

Positively Skewed Normal distribution : It is a normal distribution in which more data are clustered around the left tail side and the right tail are more flatten . It is also called right-skewed distribution. In this case mode< median < mean .

Negatively Skewed Normal distribution :It is a normal distribution in which more data are clustered around the right tail side and left tail are more flatten . It is also called left-skewed distribution. In this case mean< median < mode .
